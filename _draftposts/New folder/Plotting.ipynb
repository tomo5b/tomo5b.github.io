```python
import pandas as np
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import seaborn as sns
```

```python
# Notebook Config
np.random.seed(24)
plt.style.use('C:/Users/tomos/Documents/rose-pine-matplotlib-main/themes/rose-pine-moon.mplstyle')
```

```python
samples = np.random.normal(100, 2.5, size=(100,1)) # Generating Samples From Normal Distribution
```

```python
fig, ax = plt.subplots()
sns.histplot(samples, bins=8, color='orange', ax=ax, legend=False)
ax.set_ylabel("Count of Failures")
ax.set_xlabel("Time of Failure (Hours)")
```

```python
fig, ax = plt.subplots()
sns.histplot(samples, bins=8, color='orange', ax=ax, legend=False, stat='percent')
ax.set_ylabel("Count of Failures")
ax.set_xlabel("Time of Failure (Hours)")
```

```python
fig, ax = plt.subplots()
sns.kdeplot(samples, color='orange', legend=False)
ax.set_ylabel("Count of Failures")
ax.set_xlabel("Time of Failure (Hours)")
ax.set_ylabel('$f(x)$')
```

```python
fig, ax = plt.subplots()
sns.kdeplot(samples, ax=ax, legend=False)
x = ax.lines[-1].get_xdata()
y = ax.lines[-1].get_ydata()
x_special = 98
ax.fill_between(x, 0, y, where=x < x_special, color='gold', alpha=0.2)
ax.fill_between(x, 0, y, where=x > x_special, color='green', alpha=0.2)

ax.set_ylabel('$f(x)$')
ax.set_xlabel("Time of Failure (Hours)")
ax.annotate('$F(x)$', xy=(95, 0.02),xytext=(92.5, 0.06), arrowprops=dict(facecolor='white', shrink=0.05),
            horizontalalignment='left',
            verticalalignment='bottom',
            clip_on=True)
ax.annotate('$R(x)$', xy=(102, 0.04),xytext=(107.5, 0.08), arrowprops=dict(facecolor='white', shrink=0.05),
            horizontalalignment='left',
            verticalalignment='bottom',
            clip_on=True)
ax.annotate('$f(x)$', xy=(99.3, 0.12),xytext=(92.5, 0.12), arrowprops=dict(facecolor='white', shrink=0.05),
            horizontalalignment='left',
            verticalalignment='bottom',
            clip_on=True)
plt.show()  
```
